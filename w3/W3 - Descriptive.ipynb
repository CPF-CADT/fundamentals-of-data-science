```python
# Exercise 8
# Calculate Variance of numerical variable and explain the result
Variance = numeric_cols[
    ["vote", "total_seat", "reg_v", "cnrp", "cnrp_seat", "cpp_seat", "cpp"]
].var()

formatted_variance = Variance.apply(lambda x: f"{x:,.0f}")
formatted_variance

# The variance of vote is large, which means the number of votes is have long distance form each other. 
# The votes are spread out and are not close to the average.

# cnrp_seat and cpp_seat have small variances, which means their values are more similar 
# and stay close to the average, with less change.
```

```python
# Exercise 9
# Calculate Standard Deviation of numerical variable and explain the result
standard_deviation = Variance ** 0.5
standard_deviation
# Variables related to votes have large standard deviations, showing high data spread, 
# while seat-related variables have small standard deviations, indicating stable and consistent values.
```

```python
# Exercise 11
# plot Boxplot and explain result

import matplotlib.pyplot as plt

numeric_cols = df.select_dtypes(include='number').drop(columns=["_id", "pro_code"], errors='ignore')

cols = ["vote", "total_seat", "reg_v", "cnrp", "cnrp_seat", "cpp_seat", "cpp"]

plt.figure(figsize=(10, 6))
numeric_cols[cols].boxplot()
plt.title('Boxplots of numerical variables')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# - Boxplots show medians and IQR; dots outside whiskers are potential outliers.  
# - Vote-related variables show wide boxes and several outliers; seat-related variables are concentrated with few extremes.
```

```python
# Exercise 12
# calculate IQR and explain

cols = ["vote", "total_seat", "reg_v", "cnrp", "cnrp_seat", "cpp_seat", "cpp"]
iqr = (numeric_cols[cols].quantile(0.75) - numeric_cols[cols].quantile(0.25)).round(0)
print('IQR (per col):')
print(iqr)

# Votes (vote): 326,900 → big differences in votes between provinces.
# Total seats (total_seat): 6 → most provinces have similar seat counts.
# Registered voters (reg_v): 482,450 → some provinces have many more voters than others.
# CNRP votes (cnrp): 151,458 → moderate variation in votes for CNRP.
# CNRP seats (cnrp_seat): 3 → seat distribution for CNRP is fairly consistent.
# CPP seats (cpp_seat): 3 → seat distribution for CPP is also consistent.
# CPP votes (cpp): 154,957 → moderate variation in votes for CPP.
```

```python
# Exercise 13
# calculate lower_bound of outlier and upper_bound of outlier and explain the result 

cols = ["vote", "total_seat", "reg_v", "cnrp", "cnrp_seat", "cpp_seat", "cpp"]
iqr = numeric_cols[cols].quantile(0.75)-numeric_cols[cols].quantile(0.25)
mask = ((numeric_cols[cols] < (numeric_cols[cols].quantile(0.25)-1.5*iqr)) | (numeric_cols[cols] > (numeric_cols[cols].quantile(0.75)+1.5*iqr)))
print('Counts:', mask.sum().to_dict())
print(df.loc[mask.any(axis=1), ['province','provin_en']+cols])

# Most columns have no missing data; only 3 columns have 1 missing value each.
# Kampong Cham: CNRP got 10 seats, CPP got 8 seats out of 18.
```

```python
# Exercise 17
# Calculate Skewness of numerical variables and explain the result

from scipy.stats import skew
numeric_cols = df.select_dtypes(include='number')
# Drop identifier columns if present
numeric_cols = numeric_cols.drop(columns=["_id", "pro_code"], errors="ignore")

cols = ["vote", "total_seat", "reg_v", "cnrp", "cnrp_seat", "cpp_seat", "cpp"]
# Use pandas' built-in skew or scipy.stats.skew; both are fine. Here we use pandas for simplicity.
skewness = numeric_cols[cols].skew()

print(skewness)

# Skew > 0 → right-skewed (long tail on higher values).
# Most columns (vote, reg_v, cnrp, total_seat, cpp_seat, cnrp_seat) are slightly to moderately right-skewed → most provinces have lower values, few have very high.
# cpp (0.56) → mild right skew, more balanced than others.
```

```python
# Exercise 19
# Calculate Kurtosis  of numerical variable and explain the result 
numeric_cols = df.select_dtypes(include='number').drop(columns=["_id", "pro_code"], errors='ignore')
cols = ["vote", "total_seat", "reg_v", "cnrp", "cnrp_seat", "cpp_seat", "cpp"]

kurtosis = numeric_cols[cols].kurt()
print(kurtosis.round(3))

# Simple interpretation (thresholds):
interpret = kurtosis.apply(lambda x: 'leptokurtic (heavy tails)' if x > 0.5 else ('platykurtic (light tails)' if x < -0.5 else 'near normal'))

print(interpret)

# Most vote-related numbers are close to normal, but seat counts (especially CNRP and CPP seats) have heavier tails, 
# meaning a few provinces have unusually high or low seat counts compared to the rest.
```
