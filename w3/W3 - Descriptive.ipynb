```python
# Exercise 3
# Calculate Minimum of numerical variable and explain the result

numeric_cols = df.select_dtypes(include="number")

numeric_cols = numeric_cols.drop(columns=["_id", "pro_code"], errors="ignore")

min_values = numeric_cols.min().sort_values()

min_values
# cpp won at least one seat per province
# party get at leat 101 vote (cnp)
# some party dont have any seat in assembly
```

```python
# Exercise 4
# Calculate Maximum of numerical variable and explain the result
max_values = numeric_cols.max().sort_values()

max_values

# cnrp get most seat is 10 in a province
# party get at most is 457819 vote (cnrp)
```

```python
# Exercise 5
# Calculate mean of numerical variable and explain the result
mean = numeric_cols[["vote", "total_seat", "reg_v", "cnrp", "cnrp_seat", "cpp_seat", "cpp"]].mean()
mean

# Based on calculations across all provinces, the average are follows:
# Each province holds an average of 5 seats in the assembly.
# From an average of 403,143 registered voters per province, only 276,131 actually cast a vote.

# Party:
# People that vote for CNRP is around 122757 and get around 2 seat per province
# People that vote for CPP is around 134832 ad get around 3 seath per province
```

```python
# Exercise 6
# Calculate Median of numerical variable and explain the result
median = numeric_cols[["vote", "total_seat", "reg_v","cnrp","cnrp_seat","cpp_seat","cpp"]].median()
median

# Based on calculations across all provinces, the median are follows:
# In the middle-ranking province, 246,230 people cast a vote out of 360,881 registered voters.
# Assembly Seats: The middle-ranking province has 4 or 5 seats in the assembly.

# Party:
# People that vote for CNRP in the middle is 80471 and get around 2 seat per province
# People that vote for CPP is around 135467 ad get around 3 seath per province
```

```python
# Exercise 7
# Calculate Mode of numerical variable and explain the result
mode = numeric_cols[["total_seat","cnrp_seat","cpp_seat"]].mode()
mode

# for vote we cannot find mode but we found mode in seat that the most common seat perprovide is 1
# Party
# CNRP is mostly get ZERO seat for represent the provice 
# CPP is mostly get atleast 1 seat for represent
```

## Total

The average and middle  numbers show similar results for voters. However, for seats, these numbers are not accurate because a few provinces have a very high number of seats. While the average and middle is 4 to 5 seats, the most common number of seats (mode) is just 1.
 
## Party
we can found that average and middle number of vote for each party is show in difference while seat is very similar to each other (2,3). but for mode is far differnce for this value is only (1,0)


```python
# Exercise 8
# Calculate Variance of numerical variable and explain the result
Variance = numeric_cols[
    ["vote", "total_seat", "reg_v", "cnrp", "cnrp_seat", "cpp_seat", "cpp"]
].var()

formatted_variance = Variance.apply(lambda x: f"{x:,.0f}")
formatted_variance

# The variance of vote is large, which means the number of votes is have long distance form each other. 
# The votes are spread out and are not close to the average.

# cnrp_seat and cpp_seat have small variances, which means their values are more similar 
# and stay close to the average, with less change.
```
